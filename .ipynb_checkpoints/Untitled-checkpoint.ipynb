```python
# let's clean the dataset a little before moving on

# 1. Remove attributes that just arent useful for us
for col in ['id','funded_amnt_inv','sub_grade','emp_title','pymnt_plan','url','title','zip_code','earliest_cr_line','mths_since_last_delinq','mths_since_last_record','pub_rec','revol_bal','revol_util','total_acc','initial_list_status','out_prncp_inv','total_pymnt_inv','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','next_pymnt_d','last_credit_pull_d','collections_12_mths_ex_med','policy_code','application_type','annual_inc_joint','dti_joint','verification_status_joint','open_acc_6m','open_il_6m','open_il_12m','open_il_24m','mths_since_rcnt_il','total_bal_il','il_util','open_rv_12m','open_rv_24m','max_bal_bc','all_util','total_rev_hi_lim','inq_fi','total_cu_tl','inq_last_12m']:
    if col in df:
        del df[col]
df.info()
```

```python
summary(df)
```

```python
df['Int_Rate_Range'] = pd.cut(df.int_rate,[0,5,10,15,20,25,30],6,labels=['VeryLow','Low','Medium', 'High', 'VeryHigh', 'Expensive']) # this creates a new variable
df.Int_Rate_Range.describe()
```

```python
# Loan Classifier tells whether the loan is Bad Loan or was Paid Off completely
df['Loan_Classifier'] = 0
df['Delinq_Classifier'] = 'NA'
```

```python
df.loc[df['loan_status'] == 'Charged Off' , 'Loan_Classifier'] = 'Bad'
df.loc[df['loan_status'] == 'Default' , 'Loan_Classifier'] = 'Bad'
df.loc[df['loan_status'] == 'Late (31-120 days)' , 'Loan_Classifier'] = 'Bad'
df.loc[df['loan_status'] == 'Late (16-30 days)' , 'Loan_Classifier'] = 'Bad'
df.loc[df['loan_status'] == 'Does not meet the credit policy. Status:Charged Off' , 'Loan_Classifier'] = 'Bad'

df.loc[df['loan_status'] == 'Fully Paid' , 'Loan_Classifier'] = 'Good'
df.loc[df['loan_status'] == 'Does not meet the credit policy. Status:Fully Paid' , 'Loan_Classifier'] = 'Good'

df.loc[df['loan_status'] == 'Current' , 'Loan_Classifier'] = 'Running'
df.loc[df['loan_status'] == 'In Grace Period' , 'Loan_Classifier'] = 'Running'
df.loc[df['loan_status'] == 'Issued' , 'Loan_Classifier'] = 'Running'
```

```python
df.loc[df['delinq_2yrs'] == 0 , 'Delinq_Classifier'] = 'NeverDelinq'
df.loc[df['delinq_2yrs'] > 0 , 'Delinq_Classifier'] = 'Delinq'
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(20, 20))
df_grouped = df.groupby(by=['grade'])
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(color='grey')
ax1.set_title(r'Loan Amount')
ax1 = (df_grouped.loan_amnt.sum()/1000000).plot(kind='barh', figsize=(10,5))
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(color='blue')
ax2.set_title(r'Interest Rate')
ax2 = df_grouped.int_rate.mean().plot(kind='barh')
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(color='red')
ax3.set_title(r'Funded Amount')
ax3 = (df_grouped.funded_amnt.sum()/1000000).plot(kind='barh')
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(color='green')
ax4.set_title(r'Avg. Annual Income')
ax4 = (df_grouped.annual_inc.mean()).plot(kind='barh')
```
